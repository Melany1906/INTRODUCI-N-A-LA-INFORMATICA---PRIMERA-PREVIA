```python
# Los datos son objetos, sobre los cuales se pueden ejecutar acciones
# Una acción (procedimiento) se conoce con el nombre de: 'método'

nombre = "ada lovelace"
nombre_="melany jaramillo"

print (nombre.title())
print (nombre_.title())

# El método title es un procedimiento (función) que se aplica sobre el dato nombre
# Puesto que el dato es una cadena (información entre comillas), el método title
# ejecuta una acción específica. En este caso, la acción consiste en convertir
# el texto en una cadena de información en donde la primera palabra se convierte en mayúscula
#
# Se puede observar que después de title (título) van dos paréntesis
# La razón de esto es porque los procedimientos (métodos) pueden recibir información
#
# En el caso de title no se requiere recibir ningún tipo de dato para hacer 
# la tarea indicada. Por tal motivo, los paréntesis están vacíos
```

```python
# Métodos para convertir a mayúsculas y minúsculas

nombre = "Ada Lovelace"
nombre_= "Melany jaramillo"

print (nombre_.upper())   # Convierte el mombre a mayúsculas
print (nombre_.lower())   # Convierte el mombre a minúsculas

print (nombre.upper())   # Convierte el mombre a mayúsculas
print (nombre.lower())   # Convierte el mombre a minúsculas
```

```python
# Uniendo cadenas (concatenación)
# Se utiliza el signo + para unir cadenas

primer_nombre = "ada"
primer__nombre = "ada"
apellidos = "lovelace"

primer___nombre= "melany"
apellidos_= "jaramillo"

nombre_completo = primer_nombre + " " + apellidos
nombre__completo = apellidos + " " + primer_nombre

nombre___completo = primer___nombre + " " + apellidos_
nombre____completo = apellidos_ + " " + primer___nombre

print (nombre___completo.title())
print (nombre____completo.title())

print (nombre__completo)
print(nombre_completo)
```

```python
# Agregando espacios en blanco (Whitespace) a cadenas (String) con Tabs (tabuladores)
# o Newlines (saltos de línea)

# En programación, los espacios en blanco se refieren a cualquier carácter no imprimible, como
# espacios, tabulaciones y símbolos de final de línea. Puede utilizar espacios en blanco para organizar
# su salida para que sea más fácil de leer para los usuarios.

# Para agregar una tabulación a su texto, use la combinación de caracteres \t

print ("Python")
print ("\tPython")
print ("\tMelany\tJaramillo")
```

```python
# Para agregar una nueva línea en una cadena, use la combinación de caracteres \n

print ("Languajes:\nPython\nC\nJavaScript\n")
print ("Nombre:Melany\nApellido:Jaramillo\nEdad:17 años")
```

```python
# Se pueden combinar saltos de línea y tabulaciones:

print ("Languajes:\n\tPython\n\tC\n\tJavaScript\n")
print ("\tNombre:Melany\n\t\tApellido:Jaramillo\n\t\t\tEdad:17 Años")
```

```python
# Eliminando espacios en blanco
# En una cadena de texto, a veces es necesario eliminar los espacios a la izquierda
# de la cadena, y en ocasiones a la derecha de la cadena
#
# Para eliminar espacios a la izquierda de una cadena se utiliza: lstrip
# Para eliminar espacios a la derecha de una cadena se utiliza: rstrip

# Eliminar espacios a la derecha
lenguaje_1 = "python   "
lenguaje_2 = "     python"

print ("Sin eliminar los espacios")
print ()
print (lenguaje_1)
print (lenguaje_2)
print ()
print ("Eliminando los espacios")
print ()
print (lenguaje_1.rstrip())
print (lenguaje_2.lstrip())
```

## Números

Los números se utilizan con bastante frecuencia en la programación para llevar la puntuación en los juegos,
reenviar datos en visualizaciones, almacenar información en aplicaciones web, etc.

Python trata los números de varias formas diferentes, dependiendo de cómo están siendo usados. Primero veamos cómo Python gestiona los números enteros, los cuales son los más sencillos de trabajar.

## Enteros

Se puede sumar (+), restar (-), multiplicar (*) y dividir (/) enteros en Python.

```python
print (2 + 3)
print (3 - 2)
print (2 * 3)
print (3 / 2)
```

```python
# Python utiliza dos asteriscos para elevar un número a una potencia

print (3 ** 2)
print (3 ** 3)
print (10 ** 6)
```

```python
# Python maneja el orden de los operadores. Observe los siguientes ejemplos

print (2 + 3*4)
print ( (2 + 3) * 4)
```

## Números de tipo float

Python llama flotante a cualquier número con un punto decimal. Este término se utiliza en la mayoría de los lenguajes de programación, y se refiere al hecho de que un punto decimal puede aparecer en cualquier posición en un número. 

Cada lenguaje de programación debe diseñarse cuidadosamente para administrar correctamente los números decimales de modo que los números se comporten  adecuadamente sin importar en dónde aparezca el punto decimal.

En la mayoría de los casos, se pueden utilizar los decimales sin sin preocuparse sobre cómo se comportan. Simplemente ingrese los números que desea utilizar, y Python seguramente hará lo que se espera con dichos números:

```python
print (0.1 + 0.1)
print (0.2 + 0.2)
print (2 * 0.1)
print (2 * 0.2)
```

Pero tenga en cuenta que a veces puede obtener una cantidad arbitraria de números decimales en su respuesta. Esto sucede en todos los lenguajes y no resulta tan crítico. Python intenta encontrar la mejor manera de representar el resultado con la mayor precisión posible, pero esto es a veces difícil dada la forma en que las computadoras tienen que representar los números internamente. En este caso, si se presenta la situación, solo ignore los lugares decimales adicionales. Ejemplos:

```python
print (0.2 + 0.1)
print (3 * 0.1)
```

```python
# Evitando errores con la función str()

# Cuando se concatenan variables tipo cadena, es a veces necesario incluir dentro de
# las cadenas valores numéricos. En estos casos lo correcto es convertir el número
# a cadena, antes de realizar la concatenación. Un número se convierte a cadena
# utilizando la función str(). Ejemplo:

edad = 24
print ("La edad de Juan es de " + str(edad) + " años")
```
